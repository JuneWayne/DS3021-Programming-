```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Job data collection.csv', header = 1)
print(df.columns)
df.head(20)
```

```python
df['Median_Pay_Hour'].round(2)
top_two = df.nlargest(2, 'Median_Pay_Hour')
colors = ['red' if occupation in top_two['Occupation'].values
          else 'blue' for occupation in df['Occupation']]
plt.bar(df['Occupation'], df['Median_Pay_Hour'], color = colors)
plt.xlabel('Occupation')
plt.ylabel('Median Hourly Pay')
plt.title('Median Hourly Pay by Occupation')
plt.xticks(rotation = 90)
plt.show()

top_two = df.nlargest(2, 'Median_Pay_Year')
colors = ['red' if occupation in top_two['Occupation'].values
          else 'blue' for occupation in df['Occupation']]
plt.bar(df['Occupation'], df['Median_Pay_Year'], color = colors)
plt.xlabel('Occupation')
plt.ylabel('Median Yearly Wage')
plt.title('Median Yearly Wage by Occupation')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.pie(df['Entry_Level_Education'].value_counts(), labels=df['Entry_Level_Education'].value_counts().index, autopct='%1.1f%%')
plt.title('Entry Level Education by Occupation')
```

```python
Year_Pay_and_Growth_Rate_correlation = df[['Median_Pay_Year', 'Job_Growth_Rate']].corr()
print(Year_Pay_and_Growth_Rate_correlation)

Year_Pay_and_Hourly_Pay_correlation = df[['Median_Pay_Year', 'Median_Pay_Hour']].corr()
print(Year_Pay_and_Hourly_Pay_correlation)
```
